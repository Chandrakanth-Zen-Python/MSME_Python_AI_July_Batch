```python
import matplotlib.pyplot as plt
```

## Line Chart

```python
# Stock market trend chart

years=[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,
       2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023]

stock_price=[
    205,208.2,207.1,209.2,210.5,202.4,212.2,210.4,215.8,230.4,258.4,265.2,256.8,
    275.6,264.5,272.1,278.5,352.2,289.5,348.5,365.2,385.4,402.1
]
```

```python
plt.figure(figsize=(8,5))
plt.plot(years,stock_price, marker = 'o')
plt.title('Stock Trend For XYZ Corporation Between 2001 & 2013')
plt.xlabel('Years')
plt.ylabel("Stock Price In INR")
plt.grid(True)
plt.show()
```

## Barchart

```python
sector = ["Automobiles",'Infrastruture','Agriculture','Tourism','IT','Education',"Textiles"]

gdp_shares = [852.65,658.423,585.45,652.42,752.45,658.24,742.52]
```

```python
plt.figure(figsize=(8,5))
plt.bar(sector,gdp_shares)
plt.title('GDP shares across different sectors')
plt.xlabel("Sector")
plt.ylabel("GDP")
plt.show()
```

## Histogram


```python
import random
```

```python
ages=[random.randint(0,100) for i in range(1000)]
```

```python
plt.figure(figsize=(10,6))
plt.hist(ages,bins=5,edgecolor='black',alpha=0.7,color='brown')


plt.title("Age Distribution In a Population")
plt.xlabel('Age')
plt.ylabel('Number Of People')

plt.show()
```

## Piechart

```python
brands = ['Apple',"Samsung","Motorola",'Redmi','Vivo','Nokia','Oppo']
market_shares = [18.3,15.2,16.4,18.2,12.5,8.7,10.7]
```

```python
plt.figure(figsize=(10,7))
plt.pie(market_shares,labels=brands,autopct='%1.1f%%',startangle=0,
        colors=["#00ff00","#0000cd","#800000","#556b2f","#bcb88a","#996515","#ffdead"])

plt.title('Smartphone Market Share in 2024')
plt.show()
```
